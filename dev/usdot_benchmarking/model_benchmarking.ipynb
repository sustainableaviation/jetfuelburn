```python
from jetfuelburn.reducedorder import sacchi_etal
```

```python
weightloss_coeff = 0.16
people_on_ozempic = 0.12
total_coeff = weightloss_coeff * people_on_ozempic
loadfactor_Montlaur = 0.835
loadfactor_Seymour = 0.819
pax_correction_factor = 1 - total_coeff

pax_max = 320
```

```python
fuel_sacchi = sacchi_etal.calculate_fuel_consumption(
    year=2025,
    pax_max=pax_max,
    pax=pax_max * loadfactor_Montlaur,
    R=7000 * ureg.km,
)
fuel_sacchi_oz = sacchi_etal.calculate_fuel_consumption(
    year=2025,
    pax_max=pax_max,
    pax=pax_max * loadfactor_Montlaur * pax_correction_factor,
    R=7000 * ureg.km,
)
fuel_sacchi
fuel_sacchi_oz
```

```python
payload_sacchi = pax_max * loadfactor_Montlaur * 100.0

payload_ozempic_sacchi = pax_max * loadfactor_Montlaur * pax_correction_factor * 100.0

abs_difference_s = fuel_sacchi - fuel_sacchi_oz

rel_difference_s = abs_difference_s / fuel_sacchi * 100.0

difference_CO2_s = abs_difference_s * 3.16
```

```python
from jetfuelburn.reducedorder import yanto_etal
```

```python
pax_nmbr = 260
weight_per_pax = 100 * ureg.kg
payload = pax_nmbr * weight_per_pax
weight_per_pax_ozempic = 87.2 * ureg.kg
payload_ozempic = pax_nmbr * (0.88 * weight_per_pax + 0.12 * weight_per_pax_ozempic)
```

```python
fuel_yanto = yanto_etal.calculate_fuel_consumption(
    acft="A333", R=7000 * ureg.km, PL=payload
)
fuel_yanto_oz = yanto_etal.calculate_fuel_consumption(
    acft="A333", R=7000 * ureg.km, PL=payload_ozempic
)
fuel_yanto

fuel_yanto_oz
```

```python
abs_difference_y = fuel_yanto - fuel_yanto_oz

rel_difference_y = abs_difference_y / fuel_yanto * 100.0

difference_CO2_y = abs_difference_y * 3.16
```
