```python
pax = 200
weight_per_pax = 100 * ureg.kg
weight_per_pax_ozempic = 85 * ureg.kg
payload = pax * weight_per_pax
payload_ozempic = pax * weight_per_pax_ozempic
```

```python
fuel = yanto_etal.calculate_fuel_consumption(
    acft='A321',
    R=2200*ureg.km,
    PL=payload
)
```

```python
fuel_ozempic = yanto_etal.calculate_fuel_consumption(
    acft='A321',
    R=2200*ureg.km,
    PL=payload_ozempic
)
```

```python
(100*ureg.kg).to('lbs')
```

```python
(fuel - fuel_ozempic) / fuel * 100
```
