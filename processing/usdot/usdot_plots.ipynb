```python
import sys

sys.path.insert(0, ".")

import plotly.graph_objects as go
from usdot import process_data_usdot_t2
```

```python
year_paths = {
    "2013": "data/T_SCHEDULE_T2_2013.csv",
    "2018": "data/T_SCHEDULE_T2_2018.csv",
    "2019": "data/T_SCHEDULE_T2_2019.csv",
    "2023": "data/T_SCHEDULE_T2_2023.csv",
    "2024": "data/T_SCHEDULE_T2_2024.csv",
    "2025": "data/T_SCHEDULE_T2_2025.csv",
}

datasets = {}
for year, path in year_paths.items():
    df = process_data_usdot_t2(
        aircraft_types_csv_path="data/L_AIRCRAFT_TYPE.csv",
        t2_csv_path=path,
    ).pint.dequantify()
    df.columns = df.columns.droplevel(1)
    datasets[year] = df
```

## Flight Movements by Aircraft Type

```python
col = "Number of flights captured"
top_types = datasets["2025"][col].sort_values(ascending=False).head(20).index.tolist()
total = sum(datasets[y][col].reindex(top_types).fillna(0) for y in datasets)
sorted_types = total.sort_values(ascending=False).index.tolist()

fig = go.Figure()
for year, df in datasets.items():
    df_filtered = df.loc[df.index.isin(top_types), col].reindex(sorted_types)
    fig.add_trace(go.Bar(name=year, x=df_filtered.index, y=df_filtered.values))

fig.update_layout(
    barmode="group",
    title="Flight Movements by Aircraft Type",
    xaxis_tickangle=-45,
    yaxis_title="Movements",
)
fig.write_image("flight_movements_by_aircraft_type.pdf")
fig.show()
```
